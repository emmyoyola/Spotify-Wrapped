# Cargar biblioteca

````python
import json

# Cargar el archivo JSON
with open(r'C:\Users\EMMY\Desktop\Spotify Account Data\YourLibrary.json', 'r', encoding='utf-8') as archivo:
    datos = json.load(archivo)

# Acceder a las secciones
tracks = datos.get('tracks', {})

library = pd.DataFrame(tracks)
````

````python
library = library.apply(lambda col: col.map(str.lower) if col.dtypes == 'object' else col)
````

````python
library.replace(',', '', inplace=True)
````

````python
library['id'] = library['track'] + '-' + library['artist']
library
````

````python
maxTime = df.groupby('id')['msPlayed'].max().reset_index()
maxTime['time_min'] = maxTime['msPlayed'].apply(lambda x: f'{x // 60000:02}:{(x % 60000) // 1000:02}')
maxTime['60%'] = maxTime['msPlayed']*0.6
maxTime
````

````python
# Canciones no reproducidas en 2024

# Realizar el merge con indicador
diferencia = library.merge(maxTime, on=['id'], how='left', indicator=True)
# Filtrar registros que están solo en df1
sinReproduccion = diferencia[diferencia['_merge'] == 'left_only'].sort_values('artist').reset_index()
print(f'{sinReproduccion.shape[0]} tracks no fueron reproducidos en el 2024')
````

````python
df = df.merge(maxTime[['id', '60%']], on= ['id'], how='left')
df
````

````python
df.shape[0]
````

````python
# Selecciono sólo los tracks en las que el tiempo de reproducción fue superior al 60%
df_60 = df[df['msPlayed'] > df['60%']]
df_60 = df_60.copy()
df_60.head()
````

````python
df_60.shape[0]
````

````python
# top_artist por tiempo escuchado

time_artist = df_60.groupby('artistName')['msPlayed'].sum().sort_values(ascending=False).reset_index()
top_artist = time_artist.nlargest(5, 'msPlayed')
top_artist
````

````python
# top_artist por # de veces escuchadas
rep_artist = df_60.groupby('artistName').size().reset_index(name='count')
top_artist2 = rep_artist.nlargest(10,'count')
top_artist2
````

````python
# Top songs por tiempo escuchado
songs = df.groupby('trackName')['msPlayed'].sum().reset_index()
top_songs = songs.nlargest(5, 'msPlayed')
top_songs
````

````python
# Top songs por número de repeticiones
rep_songs = df_60.groupby('trackName').size().reset_index(name='count')
top_songs2 = rep_songs.nlargest(10, 'count')
top_songs2
````

````python
# Nueva columna para indicar el mes
df_60['month'] = df_60['endTime'].dt.month_name()
df_60.head()
````

````python
# Canciones escuchadas por mes
songs_month = df_60.groupby('month').size().reset_index(name='count').sort_values(by='count', ascending=False)
songs_month
````

````python
min_month = df_60.groupby('month')['msPlayed'].sum().reset_index().sort_values(by='msPlayed', ascending=False)
min_month['msPlayed'] = min_month['msPlayed'].astype('int')
min_month['min_month'] = min_month['msPlayed'].apply(lambda x: f'{x // 3600000:02}:{(x % 3600000) // 60000:02}:{(x % 60000) // 1000:02}')
'''
x // 3600000: Divide el tiempo total (en milisegundos) por 3600000 (milisegundos en una hora) para obtener las horas.
(x % 3600000) // 60000: Usa el residuo de la división anterior para calcular los minutos restantes.
(x % 60000) // 1000: Usa el residuo de los minutos para calcular los segundos restantes.
'''
min_month
````

````python
# Tiempo escuchado (incluyendo todos los tracks registrados ~ incluidos aquiellos en los que se escuchó menos del 60% de la canción)
tiempo = df['msPlayed'].sum()
tiempo = tiempo.astype('int')

formato = (lambda x: f'{x // 3600000:02}:{(x % 3600000) // 60000:02}:{(x % 60000) // 1000:02}')(tiempo)
formato
````

````python
# Tiempo escuchado en días

formato2 = (lambda x: f'{x // 86400000} días, {(x % 86400000) // 3600000:02}:{(x % 3600000) // 60000:02}:{(x % 60000) // 1000:02}')(tiempo)
'''
x // 86400000: Divide el tiempo total (en milisegundos) por 86400000 (milisegundos en un día) para obtener los días.
(x % 86400000) // 3600000: Usa el residuo de los días para calcular las horas restantes.
(x % 3600000) // 60000: Calcula los minutos restantes de las horas.
(x % 60000) // 1000: Calcula los segundos restantes de los minutos.
'''
formato2
````

````python
df['fecha'] = df['endTime'].dt.date

# Día en el que escuché más tiempo 
timeDate = df.groupby('fecha')['msPlayed'].sum().reset_index()
day = timeDate.nlargest(1,'msPlayed')
day['hh'] = day['msPlayed'].apply(lambda x: f'{x // 3600000:02}:{(x % 3600000) // 60000:02}:{(x % 60000) // 1000:02}')
day
````

### Tiempo promedio escuchado por día

````python
avgTime = timeDate['msPlayed'].mean()
avgTime = avgTime.astype(int)
avgTimeDay = (lambda x: f'{x // 3600000:02}:{(x % 3600000) // 60000:02}:{(x % 60000) // 1000:02}')(avgTime)
print(f'En promedio escucho {avgTimeDay} horas diarias')
````
